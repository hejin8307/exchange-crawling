```python
from bs4 import BeautifulSoup
from time import sleep
import requests
import csv
import datetime

URL = 'http://finance.naver.com/marketindex/exchangeDailyQuote.nhn?marketindexCd=FX_'
keys = ['USDKRW', 'JPYKRW', 'EURKRW', 'CNYKRW', 'CADKRW', 'AUDKRW', 'TWDKRW', 'HKDKRW', 'MYRKRW', 'SGDKRW', 'PHPKRW', 'MXNKRW', 'CHFKRW', 'NZDKRW', 'SEKKRW', 'SARKRW']

def country():
    country_name_list = []
    for key in keys:
        country_url = 'https://finance.naver.com/marketindex/exchangeDetail.nhn?marketindexCd=FX_'
        tempUrl = country_url + key
        req = requests.get(tempUrl)
        soup = BeautifulSoup(req.text, "html.parser")
        country_names = soup.select('#container > div.h_company > h2 ')
        for country_name in country_names:
            country_name = country_name.text.strip()
            country_name_list.append(country_name)
    return country_name_list
        
def toCSV(whole_list):
    dt = datetime.datetime.now()
    file = open('exchange'+dt.strftime('%Y_%m_%d')+'.csv', 'w', encoding='UTF-8', newline='')
    csvfile = csv.writer(file)
    csvfile.writerow(['country', 'date', 'exchange', 'buying_cash', 'selling_cash', 'sending_money', 'getting_money', 'buying_TC', 'selling_check'])
    for row in whole_list:
        csvfile.writerow(row)
    file.close()
    
if __name__ == "__main__":

    country_name = country()
    
    for i in keys:
        for j in range(1,3):
            url = URL + str(i) + '&page=' + str(j)
            req = requests.get(url)
            soup = BeautifulSoup(req.text, "html.parser")
            sleep(1)
            exchange_list = []
            tbody = soup.find('tbody').find_all('tr')
            for z in tbody:
                temp = []
                temp.append(country())
                temp.append(z.find('td', {'class' : 'date'}).text)
                temp.append(z.find('td', {'class' : 'num'}).text)
                for buy_cash in z.select('td:nth-child(4)'):
                    temp.append(buy_cash.text)
                for sell_cash in z.select('td:nth-child(5)'):
                    temp.append(sell_cash.text)
                for send_money in z.select('td:nth-child(6)'):
                    temp.append(send_money.text)
                for get_money in z.select('td:nth-child(7)'):
                    temp.append(get_money.text)
                for buy_tc in z.select('td:nth-child(8)'):
                    temp.append(buy_tc.text.strip())
                for sell_check in z.select('td:nth-child(9)'):
                    temp.append(sell_check.text.strip())
                exchange_list.append(temp)
            
    toCSV(exchange_list)
```
