```python
from bs4 import BeautifulSoup
from time import sleep
import requests
import csv
import datetime

URL = 'http://finance.naver.com/marketindex/exchangeDailyQuote.nhn?marketindexCd=FX_'
keys = ['USDKRW', 'JPYKRW', 'EURKRW', 'CNYKRW', 'CADKRW', 'AUDKRW', 'TWDKRW', 'HKDKRW', 'MYRKRW', 'SGDKRW', 'PHPKRW', 'MXNKRW', 'CHFKRW', 'NZDKRW', 'SEKKRW', 'SARKRW']

def country():
    country_name_list = []
    for key in keys:
        country_url = 'https://finance.naver.com/marketindex/exchangeDetail.nhn?marketindexCd=FX_'
        tempUrl = country_url + key
        req = requests.get(tempUrl)
        soup = BeautifulSoup(req.text, "html.parser")
        country_names = soup.select('#container > div.h_company > h2 ')
        for country_name in country_names:
            country_name = country_name.text.strip()
            country_name_list.append(country_name)
    return country_name_list

def get_exchange(country_name):
    country()
    exchange_list = []
    tr_list = soup.select('body > div > table > tbody > tr')
    for tr in tr_list:
        date = tr.find('td', {'class' : 'date'}).text
        exchange = tr.find('td', {'class' : 'num'}).text
        for buy_cash in tr.select('td:nth-child(4)'):
            buy_cash = buy_cash.text
        for sell_cash in tr.select('td:nth-child(5)'):
            sell_cash = sell_cash.text
        for send_money in tr.select('td:nth-child(6)'):
            send_money = send_money.text
        for get_money in tr.select('td:nth-child(7)'):
            get_money = get_money.text
        for buy_tc in tr.select('td:nth-child(8)'):
            buy_tc = buy_tc.text.strip()
        for sell_check in tr.select('td:nth-child(9)'):
            sell_check = sell_check.text.strip()
        exchange_list.append([country_name, date, exchange, buy_cash, sell_cash, send_money, get_money, buy_tc, sell_check])
    return exchange_list[0]

def toCSV(whole_list):
    dt = datetime.datetime.now()
    if whole_list == usd_whole_list :
        file = open('exchange'+dt.strftime('%Y_%m_%d')+'.csv', 'w', encoding='UTF-8', newline='')
        csvfile = csv.writer(file)
        csvfile.writerow(['country', 'date', 'exchange', 'buying_cash', 'selling_cash', 'sending_money', 'getting_money', 'buying_TC', 'selling_check'])
        csvfile.writerow(whole_list)
    else :
        file = open('exchange'+dt.strftime('%Y_%m_%d')+'.csv', 'a', encoding='UTF-8', newline='')
        csvfile = csv.writer(file)
        csvfile.writerow(whole_list)
    file.close()

if __name__ == "__main__":
    
    country_name = country()

    usd_whole_list, jpy_whole_list, eur_whole_list, cny_whole_list, cad_whole_list, aud_whole_list, twd_whole_list, hkd_whole_list, myr_whole_list, sgd_whole_list, php_whole_list, mxn_whole_list, chf_whole_list, nzd_whole_list, sek_whole_list, sar_whole_list = [],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]
    
    for i in keys:
        url = URL + str(i) + '&page1'
        req = requests.get(url)
        soup = BeautifulSoup(req.text, "html.parser")
        if i == 'USDKRW':
            usd_whole_list = get_exchange(country_name[0])
        if i == 'JPYKRW':
            jpy_whole_list = get_exchange(country_name[1])
        if i == 'EURKRW':
            eur_whole_list = get_exchange(country_name[2])
        if i == 'CNYKRW':
            cny_whole_list = get_exchange(country_name[3])
        if i == 'CADKRW':
            cad_whole_list = get_exchange(country_name[4])
        if i == 'AUDKRW':
            aud_whole_list = get_exchange(country_name[5])
        if i == 'TWDKRW':
            twd_whole_list += get_exchange(country_name[6])
        if i == 'HKDKRW':
            hkd_whole_list += get_exchange(country_name[7])
        if i == 'MYRKRW':
            myr_whole_list += get_exchange(country_name[8])
        if i == 'SGDKRW':
            sgd_whole_list += get_exchange(country_name[9])
        if i == 'PHPKRW':
            php_whole_list += get_exchange(country_name[10])
        if i == 'MXNKRW':
            mxn_whole_list += get_exchange(country_name[11])
        if i == 'CHFKRW':
            chf_whole_list += get_exchange(country_name[12])
        if i == 'NZDKRW':
            nzd_whole_list += get_exchange(country_name[13])
        if i == 'SEKKRW':
            sek_whole_list += get_exchange(country_name[14])
        if i == 'SARKRW':
            sar_whole_list += get_exchange(country_name[15])
            
    toCSV(usd_whole_list)
    toCSV(jpy_whole_list)
    toCSV(eur_whole_list)
    toCSV(cny_whole_list)
    toCSV(cad_whole_list)
    toCSV(aud_whole_list)
    toCSV(twd_whole_list)
    toCSV(hkd_whole_list)
    toCSV(myr_whole_list)
    toCSV(sgd_whole_list)
    toCSV(php_whole_list)
    toCSV(mxn_whole_list)
    toCSV(chf_whole_list)
    toCSV(nzd_whole_list)
    toCSV(sek_whole_list)
    toCSV(sar_whole_list)
```
