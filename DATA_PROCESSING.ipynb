```python
import csv
import datetime
import re

dt = datetime.datetime.now()
with open ('exchange' +dt.strftime('%Y_%m_%d')+ '.csv', 'r', encoding='UTF8') as file:
    reader = csv.reader(file)
    lines = []
    for line in reader:
        if line[0] == '미국':
            line[0] = 'USA'
        if line[0] == '일본':
            line[0] = 'JAPAN'
        if line[0] == '유럽연합':
            line[0] = 'EUROPE'
        if line[0] == '중국':
            line[0] = 'CHINA'
        if line[0] == '캐나다':
            line[0] = 'CANADA'
        if line[0] == '호주':
            line[0] = 'AUSTRALIA'
        if line[0] == '대만':
            line[0] = 'TAIWAN'
        if line[0] == '홍콩':
            line[0] = 'HONGKONG'
        if line[0] == '말레이시아':
            line[0] = 'MALAYSIA'
        if line[0] == '싱가포르':
            line[0] = 'SINGAPORE'
        if line[0] == '필리핀':
            line[0] = 'PHILIPPINES'
        if line[0] == '멕시코':
            line[0] = 'MEXICO'
        if line[0] == '스위스':
            line[0] = 'SWITZERLAND'
        if line[0] == '뉴질랜드':
            line[0] = 'NEWZEALAND'
        if line[0] == '스웨덴':
            line[0] = 'SWEDEN'
        if line[0] == '사우디아라비아':
            line[0] = 'SAUDIARABIA'
        
        if line[2] != 'exchange':
            line[2] = re.sub(',', '', line[2])
        if line[3] != 'buying_cash':
            line[3] = re.sub(',', '', line[3])
        if line[4] != 'selling_cash':
            line[4] = re.sub(',', '', line[4])
        if line[5] != 'sending_money':
            line[5] = re.sub(',', '', line[5])
        if line[6] != 'getting_money':
            line[6] = re.sub(',', '', line[6])
        if line[7] != 'buying_TC':
            line[7] = re.sub(',', '', line[7])
        if line[8] != 'selling_check':
            line[8] = re.sub(',', '', line[8])
        
        lines.append(line)

with open('exchange' +dt.strftime('%Y_%m_%d')+ '.csv', 'w', newline='', encoding='UTF-8') as file:
    writer = csv.writer(file)
    writer.writerows(lines)
    file.close()

```
